```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
path = '../data/quotes/rvi.csv'
rvi = pd.read_csv(path)
rvi.head()
```

```python
rvi = rvi.rename({'Дата' : 'date',
            'Цена' : 'price',
            'Откр.' : 'open',
            'Макс.' : 'max',
            'Мин.' : 'min',
            'Объём' : 'volume',
            'Изм. %' : 'delta_perc',
           }, axis = 1)
rvi.head(3)
```

```python
rvi = rvi.drop('volume', axis = 1)
```

```python
def price_converter(value):
    return float(str(value).replace(',', '.'))
    
    
def delta_converter(value):
    return float(str(value).replace(',', '.').replace('%', ''))/100

rvi['date'] = pd.to_datetime(rvi['date'])
rvi['price'] = rvi['price'].apply(price_converter)
rvi['open'] = rvi['open'].apply(price_converter)
rvi['max'] = rvi['max'].apply(price_converter)
rvi['min'] = rvi['min'].apply(price_converter)
rvi['delta_perc'] = rvi['delta_perc'].apply(delta_converter)
```

```python
rvi.head(3)
```

```python
rvi['delta_open'] = rvi['open'].diff(-1)
rvi['delta_price'] = rvi['price'].diff(-1)
rvi
```

```python
path = '../data/quotes/rvi_prep.csv'
rvi.to_csv(path)
```
