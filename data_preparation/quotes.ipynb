```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
path = '../data/quotes/quots.csv'
quotes = pd.read_csv(path)
quotes.head()
```

```python
quotes.columns
```

```python
quotes = quotes.rename({'Тикер':'ticker',
              'Название':'name_ru',
              'Дата\nначала\nторгов на\nМосБирже':'ipo_dt',
              'Цена\nразмещения':'ipo_price',
              'Динамика\nв первый\nдень\nторгов':'1d_price',
              '30 дней':'1m_price',
              '90\nдней':'3m_price',
              '180\nдней':'6m_price'}, axis=1)
```

```python
quotes = quotes[['ticker', 'name_ru', 'ipo_dt', 'ipo_price', '1d_price',
       '1m_price', '3m_price', '6m_price']].iloc[:-4, :]
quotes.head()
```

```python
quotes['ipo_dt'] = pd.to_datetime(quotes['ipo_dt'])
quotes.head()
```

```python
import numpy as np
quotes = quotes.replace(to_replace='—', value=np.nan)
```

```python
for col in ['1d_price', '1m_price',
       '3m_price', '6m_price']:
    quotes[col] = quotes[col].apply(lambda x: float(str(x).replace('%', '').replace(',', '.'))/100)
```

```python
quotes['ipo_price'] = quotes['ipo_price'].apply(lambda x: float(str(x).replace('%', '').replace(',', '.')))
```

```python
quotes
```

```python
path = '../data/quotes/quots_prep.csv'
quotes.to_csv(path)
```
